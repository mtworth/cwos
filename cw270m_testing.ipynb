```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="maxwellt/commentworks_os")


detection = """Task: theme detection

Identify the main themes from these comments:

The doctor took time to explain my diagnosis in plain language | Waited 45 minutes past my appointment, super frustrating | Nurse was kind but the front desk lost my paperwork again | Follow up plan was clear and I felt heard | Quick in and out, but felt a bit rushed | Billing was confusing and I got an unexpected lab fee | The clinic was spotless, smelled like disinfectant, good sign | Doctor listened to my concerns and adjusted meds carefully | Hard to get an appointment, phone line always busy | Blood draw was painless, tech was great | Parking is a nightmare near the office | They called me the next day to check on side effects, appreciated that"""

tagging = """Task: theme tagging

Comment: Dr. Lee explained my MRI results in plain language, but the front desk kept me waiting 40 minutes past my appointment.

Possible themes: bedside manner, clear explanations, wait times, staff professionalism, billing/insurance

Select the relevant themes:"""
```

```python
messages = [
    {"role": "user", "content": detection}]

pipe(messages)
```

```python
messages = [
    {"role": "user", "content": tagging}]

pipe(messages)
```
